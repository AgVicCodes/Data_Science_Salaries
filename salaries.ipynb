```python
# Remapping by Grouping 

ranges = [0, 50000, 100000, 150000, 200000, 250000, 300000, 350000, 400000, 450000, 500000, np.inf] # bins
labels = ["0 - 50,000", "50,000 - 100,000", "100,000 - 150,000", "150,000 - 200,000", "200,000 - 250,000", "250,000 - 300,000", "300,000 - 350,000", "350,000 - 400,000", "400,000 - 450,000", "450,000 - 500,000", "500,000+"] # label

sal_df["income_range"] = pd.cut(sal_df["salary_in_usd"], bins = ranges, labels = labels)

sal_df[["salary", "salary_in_usd", "income_range"]]
```
